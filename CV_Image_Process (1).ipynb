Part 2:

```python
#import important lib ...
import cv2
from skimage import io,feature,color,filters
import numpy as np
import matplotlib.pyplot as plt
```

Part 3:

```python
img_cv=cv2.imread('/content/Screenshot 2025-07-27 141520.png')
#the output will be BGR instead of RGB ..
```

```python
#we will formaulate it using cvtColor using this function COLOR_BGR2RGB ..
img_rgb=cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)
```

```python
img_sk=io.imread('/content/Screenshot 2025-07-27 141520.png')
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Open CV Image RGB ")
plt.axis('off')
plt.imshow(img_rgb)

plt.subplot(1,2,2)
plt.title("scikit-image Image (RGB)")
plt.axis('off')
plt.imshow(img_sk)
plt.show()
```

Part 4: Convert to gray scale .

```python
img_gray=cv2.cvtColor(img_cv,cv2.COLOR_BGR2GRAY)
```

```python
gray_sk=color.rgb2gray(img_sk[:,:,:3])
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Open CV Image RGB ")
plt.axis('off')
plt.imshow(img_gray,cmap='gray')

plt.subplot(1,2,2)
plt.title("scikit-image Image (RGB)")
plt.axis('off')
plt.imshow(gray_sk,cmap='gray')
plt.show()
```

```python
#(src,kernel_size,sigma value)
blur_cv=cv2.GaussianBlur(img_gray,(5,5),0)
```

```python
blur_sk=filters.gaussian(gray_sk,sigma=1)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Open CV Image using Gaussian Blur")
plt.axis("off")
plt.imshow(blur_cv,cmap='gray')

plt.subplot(1,2,2)
plt.title("Scikit Image Using Gaussian Filter")
plt.axis('off')
plt.imshow(blur_sk,cmap='gray')

plt.show()
```

Part 5 : Canny Edge Detection

```python
#100 represent min threshold
#200 represent max threshold
edge_cv=cv2.Canny(img_cv,100,200)

edge_sk=feature.canny(gray_sk,sigma=1)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Open CV Image using Canny Edge Detection")
plt.axis("off")
plt.imshow(edge_cv,cmap='gray')

plt.subplot(1,2,2)
plt.title("Scikit Image using Canny Edge Detection")
plt.axis("off")
plt.imshow(edge_sk,cmap='gray')

plt.show()
```

Part 7:

```python
#127 ->threshold thats mean that every number up 127 mean that will be colored in white other wise will be colored black ..
#255 -> the number of white
ret, binary_cv = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(binary_cv, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

image_contour=img_rgb.copy()
#-1 -> if you put -1 that means will be draw all boundaries in the list , if we put 0 or 1 that means that will be draw just one bounday in the list
#2 ->means the thickness of the line ..
#(255,0,0) ->(Blue, Green, Red)
cv2.drawContours(image_contour, contours, -1, (255,0, 0), 2)
plt.title("Contours Detected using OpenCV")
plt.axis('off')
plt.imshow(image_contour)
```
